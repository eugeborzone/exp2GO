```python
path = 'exp2GO/data/'
```

```python
espinoza = pd.read_csv(path+'pone.0014101.s005.csv', skiprows=[0,2], index_col=1)
display(espinoza.columns)
display(espinoza.head())
```

```python
# remove genes with locus 'no_match'
display(espinoza.shape)
if 'no_match' in espinoza.index:
  display(espinoza.loc['no_match'])
  espinoza = espinoza.drop('no_match')
display(espinoza.shape) 
```

```python
espinozaDC = espinoza.loc[(espinoza['Diurnal cycles F']=='TRUE') &\
                          (espinoza['Cold responsive H']=='TRUE')].iloc[:,2:2+32]
display(espinozaDC.shape)
```

```python
espinozaDC.to_csv(path+'ara_expr_espinoza_diurnal_cold_responsive.csv')
```

```python
#dist_name = 'euclidean'
#dist_name = 'seuclidean'
dist_name = 'cosine'
#dist_name = 'correlation'
d = sd.pdist(espinozaDC, dist_name)
```

```python
ds = sd.squareform(d)
espinoza_dist = pd.DataFrame(ds, index=espinozaDC.index, columns=espinozaDC.index)
display(espinoza_dist)
```

```python
espinoza_dist.to_csv(path+'ara_expression_dist_'+dist_name+'.csv.zip')
```

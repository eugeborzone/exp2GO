```python
path = 'exp2GO/data/'
```

```python
line1 = pd.read_csv(path+'dicty.gene_expression.csv.zip', nrows=0, index_col=None)
```

```python
len(list(line1))
```

```python
zitnik = pd.read_csv(path+'dicty.gene_expression.csv.zip', skiprows=[0], index_col=None)
zitnik = zitnik.set_index([list(line1.columns)])
display(zitnik.shape)
display(zitnik.index.values)
display(zitnik.columns)
display(zitnik.head())
```

```python
zitnik = zitnik.loc[zitnik.sum(axis=1)>0]
display(zitnik.shape)
```

```python
#dist_name = 'euclidean'
#dist_name = 'seuclidean'
dist_name = 'cosine'
#dist_name = 'correlation'
d = sd.pdist(zitnik, dist_name)
```

```python
ds = sd.squareform(d)
zitnik_dist = pd.DataFrame(ds, index=zitnik.index, columns=zitnik.index)
display(zitnik_dist)
```

```python
zitnik_dist.to_csv(path+'dicty_expression_dist_'+dist_name+'.csv')
```
